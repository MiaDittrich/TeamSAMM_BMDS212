Mutations are specified in three-letter HGVS-style notation (e.g. `Ser1841Arg`). For each mutation, the notebook:
1. Verifies the WT residue at the given position matches expectation,
2. Applies the substitution,
3. Prints a windowed alignment around the mutation site with the changed position marked.

## 2. Parse and apply mutations

```python
import re

THREE_TO_ONE = {
    'Ala': 'A', 'Arg': 'R', 'Asn': 'N', 'Asp': 'D', 'Cys': 'C',
    'Gln': 'Q', 'Glu': 'E', 'Gly': 'G', 'His': 'H', 'Ile': 'I',
    'Leu': 'L', 'Lys': 'K', 'Met': 'M', 'Phe': 'F', 'Pro': 'P',
    'Ser': 'S', 'Thr': 'T', 'Trp': 'W', 'Tyr': 'Y', 'Val': 'V',
}

MUT_RE = re.compile(r'^([A-Z][a-z]{2})(\d+)([A-Z][a-z]{2})$')

def parse_mutation(notation):
    """Parse 'Ser1841Arg' -> ('S', 1841, 'R')."""
    m = MUT_RE.match(notation)
    if not m:
        raise ValueError(f"Could not parse mutation: {notation!r}")
    wt3, pos, mut3 = m.group(1), int(m.group(2)), m.group(3)
    if wt3 not in THREE_TO_ONE or mut3 not in THREE_TO_ONE:
        raise ValueError(f"Unknown amino acid in: {notation!r}")
    return THREE_TO_ONE[wt3], pos, THREE_TO_ONE[mut3]

def apply_mutation(seq, notation):
    """Return (mutant_seq, wt_aa, pos, mut_aa, matched_expected).

    `matched_expected` is True if the residue at `pos` in `seq` matches the WT
    letter in the notation. The substitution is applied regardless so you can
    still inspect the result; check the flag in the report.
    """
    wt_aa, pos, mut_aa = parse_mutation(notation)
    if pos < 1 or pos > len(seq):
        raise ValueError(f"Position {pos} out of range (sequence is {len(seq)} aa)")
    idx = pos - 1
    actual = seq[idx]
    matched = (actual == wt_aa)
    mutant_seq = seq[:idx] + mut_aa + seq[idx+1:]
    return mutant_seq, wt_aa, pos, mut_aa, matched, actual
```

## 3. Build mutant sequences and a summary table

```python
import pandas as pd

rows = []
mutants = {}  # notation -> mutant sequence

for notation in MUTATIONS:
    mutant_seq, wt_aa, pos, mut_aa, matched, actual = apply_mutation(WT_SEQ, notation)
    mutants[notation] = mutant_seq
    rows.append({
        'mutation': notation,
        'position': pos,
        'expected_WT': wt_aa,
        'actual_in_seq': actual,
        'mut_AA': mut_aa,
        'WT_matches': matched,
        'mutant_length': len(mutant_seq),
    })

summary = pd.DataFrame(rows)
summary
```

## 4. Sanity-check alignments

For each mutation, print a window of `В±FLANK` residues around the mutated position. The position itself is marked with `^` on the line below; bracketed characters on the WT and mutant rows show the residue at the mutation site.

```python
FLANK = 15  # residues to show on each side of the mutation site

def show_alignment(wt, mut, pos, notation, flank=FLANK):
    """Print a windowed WT vs mutant alignment around position `pos` (1-indexed)."""
    idx = pos - 1
    start = max(0, idx - flank)
    end = min(len(wt), idx + flank + 1)

    wt_window = wt[start:end]
    mut_window = mut[start:end]
    local_idx = idx - start  # position of mutation within the window

    # Bracket the changed residue for readability
    wt_disp = wt_window[:local_idx] + f'[{wt_window[local_idx]}]' + wt_window[local_idx+1:]
    mut_disp = mut_window[:local_idx] + f'[{mut_window[local_idx]}]' + mut_window[local_idx+1:]
    match_line = ''.join('|' if a == b else ' ' for a, b in zip(wt_window, mut_window))
    # The match_line is one character per residue; the bracket display shifts by 2 chars
    # at the mutation site, so build a pointer line that lines up with wt_disp/mut_disp.
    pointer = ' ' * local_idx + ' ^ ' + ' ' * (len(wt_window) - local_idx - 1)

    print(f'--- {notation}  (position {pos}, window {start+1}-{end}) ---')
    print(f'WT     {start+1:>5}  {wt_disp}')
    print(f'MUT    {start+1:>5}  {mut_disp}')
    print(f'                {pointer}')
    # Count total differences across the whole protein as a paranoia check
    diffs = sum(1 for a, b in zip(wt, mut) if a != b)
    print(f'Total differences vs WT across full protein: {diffs} (expected 1)')
    print()

for notation in MUTATIONS:
    show_alignment(WT_SEQ, mutants[notation], parse_mutation(notation)[1], notation)
```

## 5. (Optional) Pairwise alignment with Biopython

If you want a formal global alignment instead of a windowed view, the cell below uses Biopython's `PairwiseAligner`. Single point mutations should produce a gap-free alignment of identical length with one mismatch at the mutated position.

```python
# Colab usually has Biopython preinstalled. Uncomment if needed:
!pip install -q biopython

from Bio import Align

def align_and_verify(wt_seq, mut_dict, notation):
    """
    Performs pairwise alignment, verifies mutation site,
    and prints the alignment window.
    """
    aligner = Align.PairwiseAligner()
    aligner.mode = 'global'
    aligner.match_score = 1
    aligner.mismatch_score = -1
    aligner.open_gap_score = -10
    aligner.extend_gap_score = -1

    mut_seq = mut_dict[notation]
    alignments = aligner.align(wt_seq, mut_seq)
    alignment = alignments[0]

    # Parse mutation info
    wt_aa_exp, pos, mut_aa_exp = parse_mutation(notation)
    idx = pos - 1

    # Verification: Check that WT and MUT only differ at the specific index
    mismatches = [i for i, (a, b) in enumerate(zip(wt_seq, mut_seq)) if a != b]

    print(f"--- Alignment for {notation} ---")
    print(f"Score: {alignment.score}")

    if mismatches == [idx]:
        print(f"Verification: SUCCESS (Only 1 mismatch found at position {pos})")
    else:
        print(f"Verification: WARNING! Unexpected differences at indices: {mismatches}")

    # Show windowed view
    lo, hi = max(0, idx - 30), min(len(wt_seq), idx + 31)
    local_idx = idx - lo

    wt_win = wt_seq[lo:hi]
    mut_win = mut_seq[lo:hi]

    wt_disp = wt_win[:local_idx] + f'[{wt_win[local_idx]}]' + wt_win[local_idx+1:]
    mut_disp = mut_win[:local_idx] + f'[{mut_win[local_idx]}]' + mut_win[local_idx+1:]

    print('Windowed view (В±30 aa):')
    print(f'WT  {wt_disp}')
    print(f'MUT {mut_disp}')
    print("\n")

# Run for all mutants
for notation in MUTATIONS:
    align_and_verify(WT_SEQ, mutants, notation)
```
